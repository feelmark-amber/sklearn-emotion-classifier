```python
# 生データを1ファイルに集約
import glob
import json
import os

directory_name = '../data/raw'
file_pathes = glob.glob(f"{directory_name}/*.json")

json_list = [] # 生データを集約するためのリストを作成

for file_path in file_pathes:
    with open(file_path, 'rb') as f:
        json_load = json.load(f)
    json_list.extend(json_load)

json_list = sorted(json_list, key=lambda k: k['id'])

dist_path = '../data/raw/all.json'
if os.path.isfile(dist_path):
    os.remove(dist_path)
with open(dist_path, 'w') as f:
    json.dump(json_list, f, ensure_ascii=False, indent=4)
```

```python
# アノテーションの個数集計
import xml.etree.ElementTree as ET
import os
import json
import glob

xml_file_path = '../data/labeling/labeling_ Interface.xml'

try:
    with open(xml_file_path, 'rb') as f:
        tree = ET.parse(f) # XMLファイルをパースしてElementTreeオブジェクトを取得

    root = tree.getroot() # ルート要素を取得
    annotation_list = [] # value属性の値を格納するための空のリストを作成

    # findallメソッドとXPathを使って、<Choices>タグの下にある全ての<Choice>タグを検索
    for choice_element in root.findall('./Choices/Choice'):
        value = choice_element.get('value')
        if value is not None:
            annotation_list.append(value)
    
    annotation_result_list = [] # アノテーション結果のリストを作成
    directory_name = '../data/annotated'
    file_pathes = glob.glob(f"{directory_name}/project*.json")
    for f in file_pathes:
       with open(json_file_path, 'rb') as f:
            json_load = json.load(f)
            for r in json_load:
                for an in r['annotations']:
                    for r in an['result']:
                        annotation_result_list.extend(r['value']['choices'])

    # アノテーションごとの個数を集計
    annotation_counts = {} # アノテーション個数のdictを作成
    for an in annotation_list:
        annotation_counts[an] = annotation_result_list.count(an)

    dist_path = '../data/annotated/counts.json'
    if os.path.isfile(dist_path):
        os.remove(dist_path)
    with open(dist_path, 'w') as f:
        json.dump(annotation_counts, f, ensure_ascii=False, indent=4)

except FileNotFoundError:
    print(f"エラー：ファイルが見つかりません - {xml_file_path}")
except ET.ParseError:
    print(f"エラー：XMLファイルの解析に失敗しました - {xml_file_path}")
```
